# 1. Notebook Description

This notebook was used to create 3D plots to highlight a possible impact of feature scaling.

---

**Imports:**

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

%pylab inline
```

# 2. Plot

```python
def randrange(n, vmin, vmax):
    return (vmax - vmin)*np.random.rand(n) + vmin

n = 50
f1 = randrange(n, 0, 0.001)
f2 = randrange(n, -10, 10)
f3 = randrange(n, -10, 10)
samples = np.array([f1, f2, f3]).T

f1 = randrange(n, 0.99, 1)
f2 = randrange(n, -10, 10)
f3 = randrange(n, -10, 10)
samples = np.concatenate([samples, np.array([f1, f2, f3]).T])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xs=samples[:n,0],
           ys=samples[:n, 1],
           zs=samples[:n, 2],
           c='r', marker='o')
ax.scatter(xs=samples[n:,0],
           ys=samples[n:, 1],
           zs=samples[n:, 2],
           c='b', marker='o')

ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.set_zlim([-10,10])
ticks = [10,5,0,-5,-10]
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_zticks(ticks)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

samples=MinMaxScaler(feature_range=(-10,10)).fit_transform(samples)

ax.scatter(xs=samples[:n, 0],
           ys=samples[:n, 1],
           zs=samples[:n, 2],
           c='r', marker='o')
ax.scatter(xs=samples[n:, 0],
           ys=samples[n:, 1],
           zs=samples[n:, 2],
           c='b', marker='o')

ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.set_zlim([-10,10])
ticks = [10,5,0,-5,-10]
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_zticks(ticks)
plt.show()
```
